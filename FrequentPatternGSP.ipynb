```python
from sys import stdin
import sys

totalCountAcrossTransactions = dict()
itemToTransactionAndIndex = dict()
transactionToItemArray = dict()
results = dict()

def incrementCounts(k,totalCountAcrossTransactions,itemToTransactionAndIndex,firstItem,secondItem,transId, indexofItInstn):   
    if secondItem is not None:
        firstItem = firstItem + " " + secondItem        
    
    if firstItem in totalCountAcrossTransactions:
        totalCountAcrossTransactions[firstItem] += 1
    else:
        totalCountAcrossTransactions[firstItem] = 1
        
    transToIndex = dict()
    transIdToIndexesWhere = dict()
    KitemToTransactionAndIndex = dict()
    listOfThem = set()
        
    if k in itemToTransactionAndIndex:
        KitemToTransactionAndIndex = itemToTransactionAndIndex[k]
        if firstItem in KitemToTransactionAndIndex:
            transIdToIndexesWhere = KitemToTransactionAndIndex[firstItem]
            if transId in transIdToIndexesWhere:
                listOfThem = transIdToIndexesWhere[transId]        
      
    listOfThem.add(indexofItInstn)
    transIdToIndexesWhere.update({transId:listOfThem})
    KitemToTransactionAndIndex.update({firstItem:transIdToIndexesWhere})
    itemToTransactionAndIndex.update({k:KitemToTransactionAndIndex})
    
def removeOldK(k,itemToTransactionAndIndex):
    if k > 2:
        itemToTransactionAndIndex.pop(k-2,None)

def updateCounts(firstItem,k, totalCountAcrossTransactions,transactionToItemArray,itemToTransactionAndIndex,kOneFrequentItems) :
    transIdToIndexesWhere = dict()
    listOfTransWithItem = set()
    KitemToTransactionAndIndex = dict()
    if k > 0 and itemToTransactionAndIndex[k-1]:
        KitemToTransactionAndIndex = itemToTransactionAndIndex[k-1]
    elif itemToTransactionAndIndex[k] :
        KitemToTransactionAndIndex = itemToTransactionAndIndex[k]

    if firstItem in KitemToTransactionAndIndex:
        transIdToIndexesWhere = KitemToTransactionAndIndex[firstItem]
        for tranxWhereItIs in transIdToIndexesWhere:
            indexesWhereItIs = transIdToIndexesWhere[tranxWhereItIs]
            sequenceOfTrans = transactionToItemArray[tranxWhereItIs]
            for whereToStart in indexesWhereItIs:
                if whereToStart+1 < len(sequenceOfTrans):
                    secondItem = sequenceOfTrans[whereToStart+1]
                    if secondItem in kOneFrequentItems:
                        incrementCounts(k,totalCountAcrossTransactions,
                        itemToTransactionAndIndex,firstItem,secondItem,
                                        tranxWhereItIs,whereToStart + 1)

        
def iterateK(k, totalCountAcrossTransactions,transactionToItemArray,itemToTransactionAndIndex,kOneFrequentItems,results):
    if(k==1):
        kFrequentItems = kOneFrequentItems
    else:
        kFrequentItems = dict((k, v) for (k, v) in totalCountAcrossTransactions.items() if v >= 2)
    
    if( k > 1):
        results.update(kFrequentItems)
    
    if k < 5 and len(kFrequentItems) != 0 :
        totalCountAcrossTransactions.clear()
        map(lambda (j,firstItem): updateCounts(firstItem,k, totalCountAcrossTransactions,transactionToItemArray,itemToTransactionAndIndex,kOneFrequentItems)
            ,enumerate(kFrequentItems))                                            
        removeOldK(k,itemToTransactionAndIndex)
        iterateK(k+1, totalCountAcrossTransactions, 
                  transactionToItemArray,itemToTransactionAndIndex,kOneFrequentItems,results)         
    
transId=-1;
for line in sys.stdin.readlines():
    transId = transId + 1
    row = line.split()
    map(lambda (j,firstItem): incrementCounts(0,totalCountAcrossTransactions,itemToTransactionAndIndex,firstItem,None,transId,j),enumerate(row))        
    transactionToItemArray.update({transId:row})

kOneFrequentItems = dict((k, v) for (k, v) in totalCountAcrossTransactions.items() if v >= 2)  

iterateK(1,totalCountAcrossTransactions, transactionToItemArray, itemToTransactionAndIndex,kOneFrequentItems,results)

results = sorted(results.items(), key = lambda y: (-y[1], y[0]))

results = results[0:20]

for seq in results:
    print("[" + str(seq[1]) + ", '" + seq[0] + "']")
```
